```python
from transformers import pipeline

# Cargamos un modelo específico para análisis de sentimientos en español
analizador = pipeline("sentiment-analysis", model="pysentimiento/robertuito-sentiment-analysis")

# Probamos con algunas frases
frases = [
    "Me encanta el nuevo curso de IA, es muy productivo.",
    "El servicio de entrega fue pésimo, nunca llegó el pedido.",
    "Mañana es la reunión en la zona de Buenos Aires."
]

resultados = analizador(frases)

for frase, res in zip(frases, resultados):
    print(f"Frase: {frase} | Resultado: {res['label']} ({res['score']:.2f})")
```

```python


from textblob import TextBlob

def analizar_sentimientos(frases):
    print(f"{'Frase':<50} | {'Sentimiento':<15} | {'Puntaje'}")
    print("-" * 80)

    for frase in frases:
        # Creamos un objeto TextBlob
        # Nota: TextBlob funciona mejor traduciendo internamente al inglés para el análisis
        blob = TextBlob(frase)

        try:
            # Traducimos para mayor precisión en el análisis de polaridad
            traduccion = blob.translate(from_lang='es', to='en')
            polaridad = traduccion.sentiment.polarity
        except:
            # Si la frase es muy corta o falla la traducción, analizamos el original
            polaridad = blob.sentiment.polarity

        # Clasificamos según el puntaje
        if polaridad > 0.1:
            resultado = "Positivo"
        elif polaridad < -0.1:
            resultado = "Negativo"
        else:
            resultado = "Neutro"

        print(f"{frase[:48]:<50} | {resultado:<15} | {polaridad:.2f}")

# 2. Lista de frases para probar
mis_frases = [
    "¡Estoy muy feliz con los resultados de mi nuevo proyecto!",
    "La atención al cliente fue un desastre y perdí mucho tiempo.",
    "El clima hoy está nublado en la ciudad.",
    "No me gusta nada cómo quedó el diseño final.",
    "Es una herramienta increíble y muy fácil de usar."
]

# 3. Ejecutamos la función
analizar_sentimientos(mis_frases)
```
