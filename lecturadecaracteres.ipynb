```python

import easyocr
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

# 2. Subir el archivo desde tu computadora
print("Por favor, sube la imagen que quieres procesar:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# 3. Inicializar el lector (configurado para español e inglés)
reader = easyocr.Reader(['es', 'en'])

# 4. Leer el texto de la imagen
print("\nProcesando imagen... esto puede tardar unos segundos.")
results = reader.readtext(file_name)

# 5. Mostrar resultados
print("\n--- Texto Detectado ---")
full_text = ""
for (bbox, text, prob) in results:
    print(f"Texto: {text} (Confianza: {prob:.2f})")
    full_text += text + " "

# 6. Análisis simple de frases
print("\n--- Análisis de Frases ---")
palabras = full_text.split()
print(f"Total de palabras detectadas: {len(palabras)}")
print(f"Texto completo unificado: \n{full_text}")
```

```python
# 1. Instalamos las librerías necesarias
!pip install easyocr

import easyocr
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow # Importante para mostrar imágenes en Colab

# 2. Subir el archivo desde tu computadora
print("Por favor, sube la imagen que quieres procesar:")
uploaded = files.upload()

if uploaded:
    file_name = list(uploaded.keys())[0]

    # --- NUEVAS LÍNEAS PARA MOSTRAR EL TEXTO ORIGINAL (IMAGEN) ---
    print("\n--- Imagen Original Subida ---")
    img = cv2.imread(file_name)
    # Redimensionamos un poco si la imagen es muy grande para que quepa en pantalla
    escala = 50 # porcentaje del tamaño original
    ancho = int(img.shape[1] * escala / 100)
    alto = int(img.shape[0] * escala / 100)
    img_resul = cv2.resize(img, (ancho, alto), interpolation = cv2.INTER_AREA)

    cv2_imshow(img_resul)
    # -------------------------------------------------------------

    # 3. Inicializar el lector (español e inglés)
    reader = easyocr.Reader(['es', 'en'])

    # 4. Leer el texto
    print("\nProcesando OCR... por favor espera.")
    results = reader.readtext(file_name)

    # 5. Mostrar resultados de texto
    print("\n--- Texto Extraído ---")
    full_text = ""
    for (bbox, text, prob) in results:
        print(f"[{round(prob, 2)}] {text}")
        full_text += text + " "

    print("\n--- Bloque de Texto Completo ---")
    print(full_text)
else:
    print("No se subió ningún archivo.")
```
