```python
import cv2
from google.colab.patches import cv2_imshow

# 1. Cargar la imagen
img = cv2.imread('/content/monalisa.jpg')
alto, ancho = img.shape[:2]

# 2. Calcular la constante phi
PHI = 1.61803398875

# 3. Calcular los puntos de división áurea
# Dividimos el ancho y alto por la proporción
punto_x = int(ancho / PHI)
punto_y = int(alto / PHI)

# Inversos (para el otro lado de la imagen)
punto_x_inv = ancho - punto_x
punto_y_inv = alto - punto_y

# 4. Dibujar las líneas sobre una copia
overlay = img.copy()
color = (0, 255, 255) # Amarillo en BGR
grosor = 2

# Líneas verticales
cv2.line(overlay, (punto_x, 0), (punto_x, alto), color, grosor)
cv2.line(overlay, (punto_x_inv, 0), (punto_x_inv, alto), color, grosor)

# Líneas horizontales
cv2.line(overlay, (0, punto_y), (ancho, punto_y), color, grosor)
cv2.line(overlay, (0, punto_y_inv), (ancho, punto_y_inv), color, grosor)

# 5. Mostrar resultado
cv2_imshow(overlay)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
