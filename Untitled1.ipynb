```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: tamaño de la casa (m²) y precio (miles de dólares)
X = np.array([50, 80, 100, 150, 200, 250, 300]).reshape(-1, 1)  # Característica
y = np.array([150, 200, 250, 300, 400, 500, 600])  # Precio

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo con los datos
modelo.fit(X, y)

# Predicción para un nuevo tamaño de casa
nuevo_tamano = np.array([[120]])  # Tamaño en m²
precio_predicho = modelo.predict(nuevo_tamano)

print(f"Precio predicho para una casa de 120 m²: {precio_predicho[0]:.2f} mil dólares")

# Visualización de los datos y la regresión
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X), color='red', label='Línea de regresión')
plt.scatter(nuevo_tamano, precio_predicho, color='green', marker='o', label='Predicción (120m²)')
plt.xlabel('Tamaño de la casa (m²)')
plt.ylabel('Precio (miles de dólares)')
plt.legend()
plt.show()
```
