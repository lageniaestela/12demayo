```python
import pyttsx3

engine = pyttsx3.init()

# Listamos las voces disponibles
voices = engine.getProperty('voices')

# Elegimos la voz femenina (normalmente el índice 1)
engine.setProperty('voice', voices[1].id)

engine.say("Ahora estoy usando una voz femenina.")
engine.runAndWait()
```
