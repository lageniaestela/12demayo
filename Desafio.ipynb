```python
import easyocr
import cv2
import requests
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image
from io import BytesIO



url_imagen ="https://marycabezarr.blogspot.com/2022/07/tipos.html"

response = requests.get(url_imagen)
img_raw = Image.open(BytesIO(response.content)).convert('RGB')
imagen_cv = cv2.cvtColor(np.array(img_raw), cv2.COLOR_RGB2BGR)


reader = easyocr.Reader(['es', 'en'], gpu=True)


print("Procesando imagen... por favor espera.")
resultados = reader.readtext(imagen_cv)

# 6. Mostrar resultados y dibujar sobre la imagen
print(f"\n--- Se encontraron {len(resultados)} fragmentos de texto ---\n")

for (bbox, texto, confianza) in resultados:
    # Imprimir texto en consola
    print(f"[{confianza:.2f}] Detectado: {texto}")

    # Dibujar en la imagen para ver dónde estaba el texto
    puntos = [tuple(map(int, p)) for p in bbox]
    cv2.polylines(imagen_cv, [np.array(puntos)], isClosed=True, color=(0, 255, 0), thickness=2)
    cv2.putText(imagen_cv, texto, puntos[0], cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

# 7. Mostrar la imagen final con las detecciones
print("\n--- Resultado Visual ---")
cv2_imshow(imagen_cv)
```
